```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your preprocessed dataset 
flight_data = pd.read_csv('flight.csv')

# Summary statistics for numerical columns
numerical_summary = flight_data.describe()

# Distribution of altitudes
plt.figure(figsize=(10, 6))
sns.histplot(flight_data['altitude'], bins=20)
plt.title('Distribution of Altitudes')
plt.xlabel('Altitude')
plt.ylabel('Frequency')
plt.show()

```

```python
correlation_matrix = flight_data.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()

# Pairplot for selected columns
sns.pairplot(data=flight_data[['altitude', 'distance']])
plt.title('Pplot of Altitude, Distance')
plt.show()
```

```python
avg_altitude_by_type = flight_data.groupby('aircraft_type')['altitude'].mean().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(data=avg_altitude_by_type, x='aircraft_type', y='altitude')
plt.title('Average Altitude by Aircraft Type')
plt.xticks(rotation=45)
plt.xlabel('Aircraft Type')
plt.ylabel('Average Altitude')
plt.show()

# Insight 3: Correlation between Altitude and Distance
plt.figure(figsize=(6, 6))
sns.scatterplot(data=flight_data, x='altitude', y='distance')
plt.title('Correlation between Altitude and Distance')
plt.xlabel('Altitude')
plt.ylabel('Distance')
plt.show()
```
